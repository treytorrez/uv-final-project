# CASAS Dataset Exploration

This notebook demonstrates loading and exploring the CASAS Human Activity Recognition dataset.
We'll use the structured pipeline from `src/ds_project/` to load, preprocess, and analyze the data.

```python
# Add src to path for imports
import sys
sys.path.append('../src')

import polars as pl
from ds_project.data.loaders import load_casas_home, get_available_homes
from ds_project.features.transforms import preprocess_casas_features
from ds_project.config import config
```

```python
# Check available homes
available_homes = get_available_homes()
print(f"Found {len(available_homes)} homes: {available_homes[:5]}...")  # Show first 5
```

```python
# Load a single home for exploration
home_id = 'csh101'
lf = load_casas_home(home_id)

print(f"Schema: {lf.schema}")
print(f"Estimated rows: {lf.select(pl.count()).collect().item()}")
```

```python
# Sample the data for quick analysis
df_sample = lf.head(1000).collect()
df_sample
```

```python
# Check activity distribution
activity_counts = lf.select("activity").collect().value_counts()
print("Activity distribution:")
print(activity_counts)
```

```python
# Apply preprocessing pipeline
lf_processed = preprocess_casas_features(lf)

# Check grouped activities
grouped_counts = lf_processed.select("activity_grouped").collect().value_counts()
print("Grouped activity distribution:")
print(grouped_counts)
```

```python
# Show preprocessing effects
print(f"Original schema columns: {len(lf.schema)}")
print(f"Processed schema columns: {len(lf_processed.schema)}")
print(f"\nNew columns: {set(lf_processed.schema.keys()) - set(lf.schema.keys())}")
```
